```python
def extract_episode(txt):
    check = False
    data = ""

    for i in txt:
        if i == '(':
            check = True
            continue   # '(' ko add nahi karna

        if i == ')':
            check = False
            break      # ')' pe stop karna

        if check:
            data = data + i

    return data

def name(txt):
    
    data = ""

    for i in txt:
        if i == '(':
            break

        else: data=data+i

    return data
```

```python
df['episode']=df['Title'].apply(extract_episode).str.split(' ').str[0]
df['Name']=df['Title'].apply(name).str.split(':').str[0]
```
